# Slicing.
- Access part of string.

- str[starting_idx : ending_idx]#ending idx not included.
- str = "apna College"
- str = [1:4] is "pna".
- str = [ : 4] is same as str [0:4]
- str = [1:] is same as str [1:len(str)]

#negative sliding :-

 A   P   P   L   E
-5  -4  -3  -2  -1

- str = "Apple"
- str [-3: -1] is "pl".

```python
str = "apna_college"
print(str[0:4])

str = "apna_college"
print(str[:4])#same as str[0:4]
print(str[5:])#same as str[5:len(str)]

#negative slicing
str = "apple"
str[-3:-1]
```

# string functions.

```python
#str.endwith() function # returns true if string end with substr otherwise false.
str1 = "apna_college"
print(str1.endswith("ege"))
print(str1.endswith("college123"))

#str.capitalize() function # capitalizes first letter of string.
str1 = "i am studying python from shradha khapra"
str1 = str1.capitalize()
print(str1)    

#str.replace(old , new) : replace all occurence of old substring with new substring.
str = " i am studying python from apna college"
print(str.replace("o","a" ))

#str.find(word) : returns 1st index of 1st occurence.
str = "i am from studying python from apna college"
print(str.find("o"))
print(str.find("python"))

# str.count("am") : counts the occurence of substring.
str = "i am from studying python from apna college"
print(str.count("o"))
```

# Conditional Statement :-
- if elif - else.

- if (Condition):-
     statement1 :- Always check for condition.

- elif (Condition).
     statement2 :- Check for it if condition gives false.

- else:
    statement N :- if both iF and elif are false the else statement is work.     

```python
#if statement.

age = 24
if(True):
    print("can vote")
    print("can drive")

#elif statement.

light = "yellow"

if(light == "red"):
    print("stop")
if(light == "green"):
    print("go")
elif(light == "yellow"):
    print("look")

print("end of code")

#else statement.
age =17
if(age >= 18):
    print("can vote")#indentation is must.(proper spacing)
else:
    print("cannot vote")
```

# Nesting :-
- means in if condition you can add one more if condition.

```python
#nesting code.
age = 56
if(age >= 18):
    if(age <= 80):
        print("cannot drive")
else:
        print("can drive")  
      
```
