# Dictionary in python.
- are used to store data values in keys : value pairs
- They are unordered , mutable(changeable) and dont allow duplicate keys.

```python
info = {
    "name" : "apnacollege",
    "Subjects":["python", "java"] ,
    "topic"  :("dict","set"),
    "age" : 35,
    "is_adult" : True,
    12.99 : 94.4
 } 
print(info)
```

```python
# Dictionary ["key"] = "Value" # to assign or add new.

info = {
    "name" : "apnacollege",
    "Subjects":["python", "java"] ,
    "topic"  :("dict","set"),
    "age" : 35,
   
 } 
info["name"]= "shreyas"
info["surname"]= "bhavar"
print(info)

#also we can make null_dictionary={}

null_dict = {}
null_dict["name"] = "shreyas"
print(null_dict)
```

# Nested Dictionary
- means :- in if condition there is one more if condition.

```python
#nested dictionary.

student ={
    "name" :"shreyas bhavar",
    "subjects": {
        "physics": 97,
        "chemistry":78,
        "math":95,

    }
    }
print(student)
print(student["subjects"]["chemistry"])#if we want we can print specific print.
```

# Dictionary method.

```python
# mydict.keys()#return all keys.

student = {
    "name" :"shreyas bhavar",
    "subjects": {
        "physics": 97,
        "chemistry":78,
        "math":95,

    }
    }
print(student.keys())
print(list(student.keys()))
```

```python
#mydict.values() #return all values.

student = {
    "name" :"shreyas bhavar",
    "subjects": {
        "physics": 97,
        "chemistry":78,
        "math":95,

    }
    }
print(student.values())
print(list(student.values()))
```

```python
#mydict.items()

student ={
    "name" :"shreyas bhavar",
    "subjects": {
        "physics": 97,
        "chemistry":78,
        "math":95,

    }
    }
print(student.items())
print(list(student.items()))
```

```python
#mydict.get("key")

student ={
    "name" :"shreyas bhavar",
    "subjects": {
        "physics": 97,
        "chemistry":78,
        "math":95,

    }
    }

print("Before")
print(student["name"])#["name2"]#error.
print("After")
```

```python
#mydict.update(newdict)#insert the specified item to dictionary.

student ={
    "name" :"shreyas bhavar",
    "subjects": {
        "physics": 97,
        "chemistry":78,
        "math":95,

    }
    }

new_dict={"name":"shreya","age":16}
student.update(new_dict)
print(student)
```

# Sets in python:-
- set is the collection of unordered items . Each elements in set must be unique and immutable.
-  num = {1,2,3,4}

- set 2 ={1,2,2,2}
- #duplicate value ignore in sets.

```python
#collection.

collection = {1,2,3,4,2,3,4}#duplicate value ignore in sets.
print(collection)
print(type(collection))
print(len(collection))#total no of items.
```

```python
#null_set = set() # empty set syntax.

collection = set()#empty set; syntax
print(type(collection))
```

# Set methods.
- set.add(el) #add an elements.
- set.remove() #remove an elements.
- set.clear() #empty the set
- set.pop() #removes a random value.
- set.union(set2) #combines both set values and return new.
- set.intersection(set2) #combines common value and returns new.

```python
#set.add(el) #add an elements.

collection = set()
collection.add(1)
collection.add(2)
collection.add(2)
print(collection)
```

```python
#set.remove()# to remove an elements.

collection = set()
collection.add(1)
collection.add(2)
collection.remove(1)
#collection.remove(7)#does not exist #error.

print(collection)

#we also add "string","tuple", but not "list".
```

```python
#set.clear() #empty set.

collection = set()
collection.add(1)
collection.add(2)
collection.add("shreyas")
collection.add((1,2,3))

collection.clear()
print(len(collection))
```

```python
# set.pop() #removes a random value.

collection = {"Hello","apnacourse","world","coding","python"}
print(collection.pop())
print(collection.pop())
```

```python
#set.union(set2)#combines both set values and return new.

set1 = {1,2,3}
set2 = {2,3,4}

print(set1.union(set2))
print(set1)
print(set2)
```

```python
#set.intersection(set2)#combines common value and return new.

set1 = {1,2,3}
set2 = {2,4,5}

print(set1.intersection(set2))
```
