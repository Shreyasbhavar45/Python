# List slicing :-
-  similar to string slicing.
- list_name[starting_idx: ending_idx]#ending index is not included.

- marks = [87,64,33,95,76]
  marks[1:4] is [64,33,95]
  marks[ :4] is [87,64,33,95]#is same as marks[0:4]
  marks[1: ] is same as marks len(marks)
  marks[-3:-1] is [33,95]


```python
# list slicing :-

marks = [87,64,33,95,76]
print(marks[1:4])#output:- [64, 33, 95]
print(marks[:4])#output:- [87, 64, 33, 95]
print(marks[1:])#output:- [64, 33,95,76]
print(marks[-3:-1])#output:- [33, 95]
```

# List method :-
- list.append()#adds one element at the end.
- list.sort()#sorts in ascending order.
- list.sort(reverse=True)#sort in descending order.
- list.reverse()#reverse the list.
- list.insert()#insert element at index.
- list.remove()#remove fisrt occurence of element.
- list.pop(idx)#remove element at index.

```python
#list.append()#adds one element at the end.

list = [1,2,3,4]
list.append(5)
print(list)#output:- [1, 2, 3, 4, 5]
```

```python
#list.sort()#sorts in ascending order.

list = [4,6,3,7,8,2]
print(list.sort())
print(list)#output:- [2, 3, 4, 6, 7, 8]
```

```python
#list.sort(reverse=True )#sort in descending order.
list = [4,6,3,7,8,2]
print(list.sort(reverse=True))
print(list)#output:- [8, 7, 6, 4, 3,2]
```

```python
#list.reverse()#reverse the list.
list = [1,2,3,4,5]
print(list.reverse())
print(list)#output:- [5, 4, 3, 2, 1]
```

```python
#list.insert()#insert element at index.
list =[1,2,3,4]
print(list.insert(3,56))#index 3 , element 56
print(list)#output:- [1, 2, 3, 56, 4]
```

```python
#list.remove()#remove fisrt occurence of element.
list = [1,2,3,4,2,5]
print(list.remove(2))#removes first 2
print(list)#output:- [1, 3, 4, 2, 5]
```

```python
#list.pop(idx)#remove element at index.
list = [1,2,3,4,5]
print(list.pop(2))#removes element at index 2
print(list)#output:- [1, 2, 4, 5]
```
